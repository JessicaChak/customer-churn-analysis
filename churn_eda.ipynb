```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set (style = "whitegrid")
```

```python
#Load dataset

df = pd.read_csv('../data/Telco_customer_churn.csv') 
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Churn Label'].value_counts()
```

```python
#Convert Total Charges to numeric 

df['Total Charges']=pd.to_numeric(df['Total Charges'],errors='coerce')
```

```python
#Check missing values

df.isnull().sum()
```

```python
#Drop missing values

df=df.dropna(subset=['Total Charges'])
```

```python
sns.countplot(x='Churn Label',data=df)
plt.title('Churn Distribution')
plt.show()

#Around 26% of customers have churned
```

```python
sns.histplot(df['Tenure Months'],bins=30)
plt.title('Tenure Distribution')
plt.show()

#Most customers are new 
```

```python

sns.countplot(x='Contract',hue='Churn Label',data=df)
plt.title('Churn by Contract Type')
plt.show()

#Month-to-month customers show significantly higher churn
```

```python
sns.boxplot(x='Churn Label',y='Monthly Charges',data=df)
plt.title('Monthly Charges vs Churn')
plt.show()

#Churned customers have slightly higher monthly charges on average
```

```python
sns.boxplot(x='Churn Label',y='Tenure Months',data=df)
plt.title('Tenure vs Churn')
plt.show()

#Churned customers have lower tenure, meaning they leave early 
```

```python
#Key insights: 
#Churn rate is relatively high (~26%)
#Month-to-month contracts have high churn rate
#New customers are more likely to leave
#Churned customers have higher monthly charges on average 
```
